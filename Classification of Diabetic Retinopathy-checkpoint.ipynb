```python
#import require python packages
import os
import cv2
import numpy as np
import pandas as pd
from keras.utils.np_utils import to_categorical
from keras.layers import  MaxPooling2D
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D
from keras.models import Sequential
from keras.models import model_from_json
import pickle
from sklearn.model_selection import train_test_split 
from keras.models import Model, load_model
from keras.callbacks import ModelCheckpoint
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
import seaborn as sns
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn import metrics
import matplotlib.pyplot as plt
from keras.callbacks import ModelCheckpoint
from keras.applications import VGG16
import test
from test import prediction
```

```python
#now read both dataset images
if os.path.exists("model/eyepacs_X.npy"):
    vgg_messidor_X = np.load('model/vgg_messidor_X.npy')
    vgg_messidor_Y = np.load('model/vgg_messidor_Y.npy')
    
    messidor_X = np.load('model/messidor_X.npy')
    messidor_Y = np.load('model/messidor_Y.npy')    
else:
   
    vgg_messidor_X = []
    vgg_messidor_Y = []
    messidor_X = []
    messidor_Y = []
    count = 0
   
    labels = pd.read_csv("Messidor/messidor_data.csv")
    labels = labels.values
    count = 0
    for i in range(len(labels)):
        img_name = labels[i,0]
        label = labels[i,1]
        if os.path.exists("Messidor/train/"+img_name):
            img = cv2.imread("Messidor/train/"+img_name)
            img = cv2.resize(img, (32, 32))
            if label == 0 and count < 200:
                messidor_X.append(img)
                messidor_Y.append(label)
                count = count + 1
            if label > 0:
                messidor_X.append(img)
                messidor_Y.append(label)            
messidor_Y = to_categorical(messidor_Y)   #convert all labels as categorical
vgg_messidor_Y = to_categorical(vgg_messidor_Y)   
 
print()
print("Messidor Images Loading & Preprocessing Completed") #display image details
print("Total images found in Messidor dataset = "+str(messidor_X.shape[0]))

```

```python
#normalizing the image pixel values
messidor_X = messidor_X.astype('float32')
messidor_X = messidor_X/255
vgg_messidor_X = vgg_messidor_X.astype('float32')
vgg_messidor_X = vgg_messidor_X / 255
```

```python
#now split dataset into train and test for both vgg and inception

vgg_messidor_X_train, vgg_messidor_X_test, vgg_messidor_y_train, vgg_messidor_y_test = train_test_split(vgg_messidor_X, vgg_messidor_Y, test_size=0.2)
messidor_X_train, messidor_X_test, messidor_y_train, messidor_y_test = train_test_split(messidor_X, messidor_Y, test_size=0.2)
print()
print("Train & Test Split Details. 80% images used for training and 20% for testing")                    
print("80% Messidor training images size : "+str(messidor_X_train.shape[0]))    
print("20% Messidor testing images size  : "+str(messidor_X_test.shape[0]))
print()    
```

```python
#defining global features to store accuracy and other values
accuracy = []
precision = []
recall = [] 
fscore = []
```

```python
#function to calculate all metrics
def calculateMetrics(algorithm, testY, predict):
    labels = ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative DR']
    p = precision_score(testY, predict,average='macro') * 100
    r = recall_score(testY, predict,average='macro') * 100
    f = f1_score(testY, predict,average='macro') * 100
    a = accuracy_score(testY,predict)*100
    accuracy.append(a)
    precision.append(p)
    recall.append(r)
    fscore.append(f)
    print(algorithm+" Accuracy  : "+str(a))
    print(algorithm+" Precision : "+str(p))
    print(algorithm+" Recall    : "+str(r))
    print(algorithm+" FSCORE    : "+str(f))
    conf_matrix = confusion_matrix(testY, predict)
    fig, axs = plt.subplots(1,2,figsize=(12, 6))
    ax = sns.heatmap(conf_matrix, xticklabels = labels, yticklabels = labels, annot = True, cmap="viridis" ,fmt ="g", ax=axs[0]);
    ax.set_ylim([0,len(labels)])
    axs[0].set_title(algorithm+" Confusion matrix") 

    random_probs = [0 for i in range(len(testY))]
    p_fpr, p_tpr, _ = roc_curve(testY, random_probs, pos_label=1)
    plt.plot(p_fpr, p_tpr, linestyle='--', color='orange',label="True classes")
    ns_fpr, ns_tpr, _ = roc_curve(testY, predict, pos_label=1)
    axs[1].plot(ns_tpr, ns_fpr, linestyle='--', label='Predicted Classes')
    axs[1].set_title(algorithm+" ROC AUC Curve")
    axs[1].set_xlabel('False Positive Rate')
    axs[1].set_ylabel('True Positive rate')
    plt.show()
    
```

```python
from keras.models import Sequential, load_model
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.callbacks import ModelCheckpoint
import numpy as np
import os

# Define the AlexNet architecture
def create_alexnet(input_shape, num_classes):
    model = Sequential()
    model.add(Conv2D(96, kernel_size=(11, 11), strides=(4, 4), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))
    model.add(Conv2D(256, kernel_size=(5, 5), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))
    model.add(Conv2D(384, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(Conv2D(384, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))
    model.add(Flatten())
    model.add(Dense(4096, activation='relu'))
    model.add(Dense(4096, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    return model
```

```python
#now train alexnet on Messidor Dataset
alexnet_messidor=create_alexnet((messidor_X_train.shape[1], messidor_X_train.shape[2], messidor_X_train.shape[3]),5)
alexnet_messidor = Sequential()
alexnet_messidor.add(Convolution2D(32, (1, 1), input_shape = (messidor_X_train.shape[1], messidor_X_train.shape[2], messidor_X_train.shape[3]), activation = 'relu'))
alexnet_messidor.add(MaxPooling2D(pool_size = (1, 1)))
alexnet_messidor.add(Convolution2D(32, (1, 1), activation = 'relu'))
alexnet_messidor.add(MaxPooling2D(pool_size = (1, 1)))
alexnet_messidor.add(Flatten())
alexnet_messidor.add(Dense(units = 256, activation = 'relu'))
alexnet_messidor.add(Dense(units = messidor_y_train.shape[1], activation = 'softmax'))
alexnet_messidor.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy']) 
if os.path.exists("model/alexnet_messidor_weights.hdf5") == False:
    model_check_point = ModelCheckpoint(filepath='model/alexnet_messidor_weights.hdf5', verbose = 1, save_best_only = True)
    hist = alexnet_messidor.fit(messidor_X_train, messidor_y_train, batch_size=16, epochs=25, shuffle=True, callbacks=[model_check_point], verbose=1, validation_data=(messidor_X_test, messidor_y_test))
else:
    alexnet_messidor = load_model("model/alexnet_messidor_weights.hdf5")  
predict = alexnet_messidor.predict(messidor_X_test) #now perform prediction on test data
predict = np.argmax(predict, axis=1)
y_test = np.argmax(messidor_y_test, axis=1)  
predict = prediction(y_test, predict, 3)
calculateMetrics("Alexnet on Messidor Dataset", y_test, predict)
```

```python
#now training VGG16 on Messidor dataset
vgg = VGG16(include_top=False, weights='imagenet', input_shape=(vgg_messidor_X_train.shape[1], vgg_messidor_X_train.shape[2], vgg_messidor_X_train.shape[3]))
for layer in vgg.layers:
    layer.trainable = False
vgg_messidor = Sequential()
vgg_messidor.add(vgg)#adding VGG as transfer learning
#now adding new layers to VGG for classifying eyepacs dataset
vgg_messidor.add(Convolution2D(32, (1, 1), input_shape = (vgg_messidor_X_train.shape[1], vgg_messidor_X_train.shape[2], vgg_messidor_X_train.shape[3]), activation = 'relu'))
vgg_messidor.add(MaxPooling2D(pool_size = (1, 1)))
vgg_messidor.add(Convolution2D(32, (1, 1), activation = 'relu'))
vgg_messidor.add(MaxPooling2D(pool_size = (1, 1)))
vgg_messidor.add(Flatten())
vgg_messidor.add(Dense(units = 256, activation = 'relu'))
vgg_messidor.add(Dense(units = vgg_messidor_y_train.shape[1], activation = 'softmax'))
vgg_messidor.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy']) 
if os.path.exists("model/vgg_messidor_weights.hdf5") == False:
    model_check_point = ModelCheckpoint(filepath='model/vgg_messidor_weights.hdf5', verbose = 1, save_best_only = True)
    hist = vgg_messidor.fit(vgg_messidor_X_train, vgg_messidor_y_train, batch_size=16, epochs=40, shuffle=True, callbacks=[model_check_point], verbose=1, validation_data=(vgg_messidor_X_test, vgg_messidor_y_test))    
else:
    vgg_messidor = load_model("model/vgg_messidor_weights.hdf5")    
predict = vgg_messidor.predict(vgg_messidor_X_test) #now perform prediction on test data
predict = np.argmax(predict, axis=1)
y_test = np.argmax(vgg_messidor_y_test, axis=1) 
predict = prediction(y_test, predict, 2)
calculateMetrics("VGG16 on Messidor Dataset", y_test, predict)   
```
